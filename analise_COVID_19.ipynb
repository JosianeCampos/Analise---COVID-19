## Análise de Dados - COVID-19 - Estado de São Paulo

Importando as bibliotecas necessárias inicialmente.

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
from datetime import date
```

Análise exploratória inicial

```python
df = pd.read_csv('dados/cases-brazil-states.csv')
df.head(20)
```

```python
df.info()
```

```python
df['data'] = pd.to_datetime(df.data)
```

```python
df.info()
```

```python
plt.figure(figsize = (20,8))
fig = sns.lineplot( x = 'data', y = 'Obitos', hue = 'estado', data = df )
fig.set_title('Óbitos por estado em função do tempo', loc = 'left', fontsize = 26)
fig.set_xlabel('Data', fontsize = 24)
fig.set_ylabel('Óbitos', fontsize = 24)
```

```python
df_sp = df[df.estado == 'SP']
```

```python
df_sp
```

```python
plt.figure(figsize = (20,10))
fig = sns.lineplot( x = 'data', y = 'Obitos', hue = 'estado', data = df_sp )
fig.set_title('Óbitos por estado em função do tempo', loc = 'left', fontsize = 26)
fig.set_xlabel('Data', fontsize = 24)
fig.set_ylabel('Óbitos', fontsize = 24)
```

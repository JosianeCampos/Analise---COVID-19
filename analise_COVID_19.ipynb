```python
fig, axes = plt.subplots(2, 3, sharex=True, figsize=(28,18))
fig.suptitle('Análise mensal de Novos Casos e Novos Óbitos', fontsize = (30))
sns.lineplot(ax=axes[0,0], x = 'data', y = 'novosCasos', data = df_sp[df_sp.mes == 4] )
sns.lineplot(ax=axes[0,1], x = 'data', y = 'novosCasos', data = df_sp[df_sp.mes == 5] )
sns.lineplot(ax=axes[0,2] ,x = 'data', y = 'novosCasos', data = df_sp[df_sp.mes == 6] )
sns.lineplot( ax=axes[1,0],x = 'data', y = 'novosObitos', data = df_sp[df_sp.mes == 4] )
sns.lineplot( ax=axes[1,1],x = 'data', y = 'novosObitos', data = df_sp[df_sp.mes == 5] )
sns.lineplot( ax=axes[1,2],x = 'data', y = 'novosObitos', data = df_sp[df_sp.mes == 6] )
axes[0,0].set_title('Novos Casos no mês de abril', fontsize = 25)
axes[0,1].set_title('Novos Casos no mês de maio', fontsize = 25)
axes[0,2].set_title('Novos Casos no mês de junho', fontsize = 25)
axes[1,0].set_title('Novos Óbitos no mês de abril', fontsize = 25)
axes[1,1].set_title('Novos Óbitos no mês de maio', fontsize = 25)
axes[1,2].set_title('Novos Óbitos no mês de junho', fontsize = 25)
```

```python
df_sp['dia_da_semana']= df_sp.data.dt.strftime('%A')
```

```python
grupo = df_sp.groupby('dia_da_semana')['novosObitos','novosCasos'].mean()
grupo
```

```python
grupo = grupo.reindex(['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday'])
```

```python
area = plt.figure(figsize = (18,6))
g1 = area.add_subplot(1,2,1)
g2 = area.add_subplot(1,2,2)
g1.bar(grupo.index, grupo.novosObitos)
g1.set_title('Média dos Novos Óbitos durante a semana')
g2.bar(grupo.index, grupo.novosCasos)
g2.set_title('Média dos Novos Casos durante a semana')
```

Com os gráficos acima podemos observar que a sazonalidade ocorre semanalmente. Essa sazonalidade ocorre pois a maior parte dos dados são lançados
ao sistema durante a semana, tendo uma queda de lançamento no domingo e na segunda, devido aos plantões médicos.


## Tendencia


Nos gráficos de Novos Óbitos em SP e Novos Casos em SP notamos sazonalidade e observamos que essa sazonalidade ocorre de 7 em 7 dias (semanalmente).
Agora, queremos analisar a tendencia desses gráficos. Será que temos um tendencia de redução ou aumento dos Novos Óbitos? Será que temos um tendencia de redução ou aumento dos Novos Casos?
Para analisar a tendencia, vamos suavizar as curvas dos gráficos já citados:

```python
plt.figure(figsize = (20,10))
fig = sns.lineplot( x = 'data', y = 'novosObitos', data = df_sp )
fig.set_title('Novos Óbitos em SP', loc = 'left', fontsize = 26)
fig.set_xlabel('Data', fontsize = 24)
fig.set_ylabel('Log Óbitos', fontsize = 24)
fig.set_yscale('log')
```

```python
plt.figure(figsize = (20,10))
fig = sns.lineplot( x = 'data', y = 'novosCasos', data = df_sp )
fig.set_title('Novos Casos em SP', loc = 'left', fontsize = 26)
fig.set_xlabel('Data', fontsize = 24)
fig.set_ylabel('Log Casos', fontsize = 24)
fig.set_yscale('log')
```

Suavizando as curvas:

```python
df_sp['mediaObitos_7'] = df_sp.novosObitos.rolling(window = 7, center = False).mean()
```

```python
df_sp['mediaObitos_14'] = df_sp.novosObitos.rolling(window = 14, center = False).mean()
```

```python
df_sp['mediaCasos_7'] = df_sp.novosCasos.rolling(window = 7, center = False).mean()
```

```python
df_sp['mediaCasos_14'] = df_sp.novosCasos.rolling(window = 14, center = False).mean()
```

```python
df_sp.sample(5)
```

```python
plt.figure(figsize = (20,10))
fig = sns.lineplot( x = 'data', y = 'mediaObitos_7', data = df_sp.dropna() )
sns.lineplot( x = 'data', y = 'mediaObitos_14', data = df_sp )
plt.legend(labels = ["Média móvel - 7 dias","Média móvel - 14 dias"])
fig.set_title('Média móvel em SP - Novos Óbitos', loc = 'left', fontsize = 26)
fig.set_xlabel('Data', fontsize = 24)
fig.set_ylabel('Log Óbitos', fontsize = 24)
fig.set_yscale('log')
plt.show
```

```python
plt.figure(figsize = (20,10))
fig = sns.lineplot( x = 'data', y = 'mediaCasos_7', data = df_sp.dropna() )
sns.lineplot( x = 'data', y = 'mediaCasos_14', data = df_sp )
plt.legend(labels = ["Média móvel - 7 dias","Média móvel - 14 dias"])
fig.set_title('Média móvel em SP- Novos Casos', loc = 'left', fontsize = 26)
fig.set_xlabel('Data', fontsize = 24)
fig.set_ylabel('Log Casos', fontsize = 24)
fig.set_yscale('log')
plt.show
```

```python
plt.figure(figsize = (20,8))
fig = sns.lineplot( x = 'data', y = 'mediaObitos_7', data = df_sp.dropna() )
sns.lineplot( x = 'data', y = 'mediaObitos_14', data = df_sp )
fig.set_title('Média móvel em SP - Novos Óbitos', loc = 'left', fontsize = 26)
fig.set_xlabel('Data', fontsize = 24)
fig.set_ylabel('Log Óbitos', fontsize = 24)
fig.set_yscale('log')
plt.bar(df_sp.data, df_sp.novosObitos, color = 'lightgray')
```

```python
plt.figure(figsize = (20,8))
fig = sns.lineplot( x = 'data', y = 'mediaCasos_7', data = df_sp.dropna() )
sns.lineplot( x = 'data', y = 'mediaCasos_14', data = df_sp )
fig.set_title('Média móvel em SP - Novos Casos', loc = 'left', fontsize = 26)
fig.set_xlabel('Data', fontsize = 24)
fig.set_ylabel('Log Casos', fontsize = 24)
fig.set_yscale('log')
plt.bar(df_sp.data, df_sp.novosCasos, color = 'lightgray')
```

Podemos notar que temos uma tendencia de aumento em Novos Casos e em Novos Óbitos. Observamos também um platô, ou seja, uma possível estabilização no número de casos e mortes causados pela COVID - 19.

## Aceleração dos dados

```python
df_sp['aceleracaoNovosCasos'] = df_sp.novosCasos.diff()
df_sp['aceleracaoObitos'] = df_sp.novosObitos.diff()
```

```python
df_sp.sample(5)
```

```python
fig, axes = plt.subplots(1, 2, sharex=True, figsize=(16,8))
sns.barplot(ax=axes[0], x = 'data', y = 'aceleracaoNovosCasos', data = df_sp)
sns.barplot(ax=axes[1], x = 'data', y = 'aceleracaoObitos', data = df_sp)
axes[0].set_title('Aceleração Novos Casos', fontsize = 25)
axes[1].set_title('Aceleraçao Novos Óbitos', fontsize = 25)
```

Podemos observar que temos acelerações positivas e acelerações negativas em ambos os gráficos. O que isso me diz? As acelerações negativas estão me dizendo que existe uma desaceleração no número de casos e óbitos, já as acelerações positivas estão me dizendo que existe uma aceleração no número de casos e óbitos.
Olhando apenas para a magnitude das barras, vemos que na maioria dos dias a aceleração é maior que a desaceleração, concluíndo que temos um aumento no número de casos e óbitos, apesar de termos dias em que há um redução nesses números.

## Autocorrelação

```python
from pandas.plotting import autocorrelation_plot
```

```python
plt.figure(figsize = (20,8))
fig = sns.lineplot( x = 'data', y = 'mediaObitos_7', data = df_sp.dropna() )
sns.lineplot( x = 'data', y = 'mediaObitos_14', data = df_sp )
fig.set_title('Média móvel em SP - Novos Óbitos', loc = 'left', fontsize = 26)
fig.set_xlabel('Data', fontsize = 24)
fig.set_ylabel('Log Óbitos', fontsize = 24)
fig.set_yscale('log')
plt.bar(df_sp.data, df_sp.novosObitos, color = 'lightgray')
```

```python
ax = plt.figure(figsize = (12,6))
ax.suptitle('Correlação dos Novos Óbitos', fontsize = 18, x=0.26, y = 0.95)
autocorrelation_plot(df_sp.novosObitos)
```

Podemos notar no gráfico de autocorrelação dos novos óbitos a sazonalidade já observada anteriormente. Vemos também uma autocorrelação inicialmente alta e positiva significando que temos uma dependênia linear forte do aumento no número de óbitos. Logo em seguida essa depêndencia linear fica fraca se tornando negativa, mostrando que temos uma taxa alta de cresimento, que com o passar do tempo fica mais próxima de um crescimento exponencial e não de um crescimento linear, como já notamos.

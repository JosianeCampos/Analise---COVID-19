## Análise exploratória inicial

```python
df = pd.read_csv('dados/cases-brazil-states.csv')
df.head(20)
```

```python
df.info()
```

```python
df['data'] = pd.to_datetime(df.data)
```

```python
df.info()
```

```python
plt.figure(figsize = (20,8))
fig = sns.lineplot( x = 'data', y = 'Obitos', hue = 'estado', data = df )
fig.set_title('Óbitos por estado em função do tempo', loc = 'left', fontsize = 26)
fig.set_xlabel('Data', fontsize = 24)
fig.set_ylabel('Óbitos', fontsize = 24)
```

```python
df_sp = df[df.estado == 'SP']
```

```python
df_sp
```

```python
plt.figure(figsize = (20,10))
fig = sns.lineplot( x = 'data', y = 'Obitos', hue = 'estado', data = df_sp )
fig.set_title('Óbitos por estado em função do tempo', loc = 'left', fontsize = 26)
fig.set_xlabel('Data', fontsize = 24)
fig.set_ylabel('Óbitos', fontsize = 24)
```

```python
plt.figure(figsize = (20,10))
fig = sns.lineplot( x = 'data', y = 'Obitos', hue = 'estado', data = df_sp )
fig.set_title('Óbitos por estado em função do tempo', loc = 'left', fontsize = 26)
fig.set_xlabel('Data', fontsize = 24)
fig.set_ylabel('Óbitos', fontsize = 24)
fig.set_yscale('log')
```

Podemos notar que os óbitos no estado de São Paulo, estão crescendo exponencialmente.

```python
df_sp = df_sp.fillna(0)
```

```python
df_sp
```

```python
plt.figure(figsize = (20,8))
fig = plt.bar(df_sp.data, df_sp.suspeitos, label = 'Suspeitos')
plt.bar(df_sp.data, df_sp.recuperados, label = 'Recuperados')
plt.bar(df_sp.data, df_sp.Obitos, label = 'Óbitos')
plt.xlabel('Data', fontsize = 18)
plt.ylabel('Óbitos, recuperados e suspeitos no estado de SP', fontsize = 18, rotation = 90)
plt.legend()
plt.show()

```

Podemos notar que o número de recuperados aumentou subitamente no mês de junho. Fazendo algumas pesquisas podemos entender o por que desse 
aumento repentino: isso ocorreu por que no mês de junho o estado de São Paulo divulgou, pela primeira vez, o número de recuperados que não precisaram de internação hospitalar. Anteriormente, a secretaria estadual da Saúde calculava apenas o total de altas hospitalares entre pacientes internados com Covid-19.
